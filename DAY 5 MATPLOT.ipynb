## Importing the required module

````python
import matplotlib.pyplot as plt
# x axis values
x = [1,2,3]
# corresponding y axis values
y = [2,4,1]
# plotting the points
plt.plot(x, y)
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
# giving a title to my graph
plt.title('My first graph!')
# function to show the plot
plt.show()
````

````python
import numpy as np
x = np.arange(0, 10, 0.1)
y = np.sin(x)
print(x)

# figures in matplotlib using figure()
plt.figure(figsize=(10, 6))  # 7 represent height and 4 represent width 
plt.plot(x, y)
plt.show()
````

## Legends, labels and titles

````python
# Ticks are the markers denoting data points on axes.
# importing libraries
import matplotlib.pyplot as plt
import numpy as np

# values of x and y axes
x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
y = [1, 4, 3, 2, 7, 6, 9, 8, 10, 5]

plt.plot(x, y, 'g')
plt.xlabel('x')
plt.ylabel('y')
# here we set the size for ticks, rotation and color value

plt.tick_params(axis="x", labelsize=18, labelrotation=-60, labelcolor="blue")
plt.tick_params(axis="y", labelsize=12, labelrotation=20, labelcolor="gray")

plt.show()
````

## Adding Legend in the graph

````python
# importing modules
import numpy as np
import matplotlib.pyplot as plt

# Y-axis values
y1 = [2, 3, 4.5]

# Y-axis values
y2 = [1, 1.5, 5]

# Function to plot
plt.plot(y1)
plt.plot(y2)

# Function add a legend
plt.legend(["blue", "green"], loc ="best")  # loc here defined where we want to place the lengend ex- lowwer right , upper right , right , best-->  there are only valid location of left is not considered
# here best represent where lengend will be best located by default 
# function to show the plot
plt.show()
````

## Histogram

````python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(170, 10, 250)
#The hist() function will read the array and produce a histogram:
#set the bin value 30
plt.hist(data);
# if we dont't pass any value in bins then bydefault it's define 10 bins 

plt.show() 
plt.hist(data, bins=30); #It is a type of bar graph. To construct a histogram, the first step is to “bin” the range of values — that is, divide the entire range of values into a series of intervals — and then count how many values fall into each interval. The bins are usually specified as consecutive, non-overlapping intervals of a variable

plt.show() 
````

````python
#import matplotlib.pyplot
import matplotlib.pyplot as plt
import numpy as np

#draw plot as fig
fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)

ax.plot(x, np.cos(x))
ax.axis('equal')
#text annotation to the plot where it indicate maximum value of the curv
ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))  # shrink is used to how much closer we want to point our arrow
#text annotation to the plot where it indicate minimum value of the curv
ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=-90"));


# angleA : starting angle of the path

# angleB : ending angle of the path

# armA : length of the starting arm

# armB : length of the ending arm

# rad : rounding radius of the edges

# connect(posA, posB)
````

## Pie Charts

````python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np


# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars)

# show plot
plt.show()
````

````python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np


# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars, explode=[0,0,0.2,0,0,0],autopct='%1.1f%%')

# show plot
plt.show()
````

## Logarithmic Axes in Matplotlib

````python
import matplotlib.pyplot as plt

# exponential function y = 10^x
data = [10**i for i in range(5)]

plt.plot(data)
````

````python
import matplotlib.pyplot as plt

# exponential function y = 10^x
data = [10**i for i in range(5)]

# convert y-axis to Logarithmic scale
plt.yscale("log")

plt.plot(data)
````

````python
import matplotlib.pyplot as plt

# exponential function x = 10^y
datax = [ 10**i for i in range(5)]
datay = [ i for i in range(5)]

#convert x-axis to Logarithmic scale
plt.xscale("log")   # y scale has taken default value if we don't defined 

plt.plot(datax,datay)
````

````python
# importing mplot3d toolkits, numpy and matplotlib
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

# syntax for 3-D projection
ax = plt.axes(projection ='3d')

# defining all 3 axes
z = np.linspace(0, 1, 1000) # 0 to 1 , 1000 values will be generated linearly 
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot')
plt.show()
````
